### **2. — Durasi Tidur**

Visualisasi pie chart digunakan untuk melihat distribusi durasi tidur responden.

### **Penjelasan**
* Pie chart menunjukkan persentase setiap kategori durasi tidur.
* Data ini digunakan untuk mengetahui kebiasaan tidur mahasiswi.
* Pola tidur dapat memengaruhi kondisi tubuh dan hormon.





```python
import pandas as pd
import plotly.express as px

# Rename kolom
df.rename(columns={
    '11. Berapa rata-rata durasi tidur Anda per hari?  ': 'Durasi Tidur'
}, inplace=True)

# Hitung jumlah setiap kategori durasi tidur
tidur_counts = df['Durasi Tidur'].value_counts().reset_index()

# Rename hasil perhitungan
tidur_counts.columns = ['Durasi Tidur', 'Jumlah']

# Urutkan kategori
urutan = ['< 5 jam', '5-6 jam', '7-8 jam', '> 8 jam']
tidur_counts['Durasi Tidur'] = pd.Categorical(
    tidur_counts['Durasi Tidur'],
    categories=urutan,
    ordered=True
)
tidur_counts = tidur_counts.sort_values('Durasi Tidur')

# Buat bar chart
fig = px.bar(
    tidur_counts,
    x='Durasi Tidur',
    y='Jumlah',
    text='Jumlah',
    color='Jumlah',
    color_continuous_scale='Blues',
    title='Distribusi Durasi Tidur'
)

# Percantik tampilan
fig.update_traces(
    textposition='outside',
    hovertemplate='<b>%{x}</b><br>Jumlah Responden: %{y}<extra></extra>'
)

fig.update_layout(
    template='plotly_white',
    title_x=0.5,
    font=dict(size=14),
    height=600,
    xaxis_title='Durasi Tidur',
    yaxis_title='Jumlah Responden'
)

fig.show()
```

###**1. — HUBUNGAN DURASI TIDUR DAN TINGKAT NYERI**

Visualisasi digunakan untuk melihat hubungan antara durasi tidur dan tingkat nyeri menstruasi.

```python
# Rapihin semua nama kolom
df.columns = df.columns.str.strip()

# Rename kolom
df = df.rename(columns={
    '4. Seberapa tinggi tingkat nyeri yang Anda rasakan saat menstruasi?': 'nyeri',
    'Durasi Tidur': 'durasi_tidur'
})

# Membuat crosstab
crosstab_tidur = pd.crosstab(
    df['nyeri'],
    df['durasi_tidur']
)

# Heatmap interaktif
fig = px.imshow(
    crosstab_tidur,
    text_auto=True,
    color_continuous_scale='Blues',
    title='Hubungan Durasi Tidur dan Tingkat Nyeri'
)

# Percantik tampilan
fig.update_layout(
    template='plotly_white',
    title_x=0.5,
    height=600,
    font=dict(size=14),
    xaxis_title='Durasi Tidur',
    yaxis_title='Tingkat Nyeri'
)

# Hover interaktif
fig.update_traces(
    hovertemplate=
    '<b>Tingkat Nyeri:</b> %{y}<br>' +
    '<b>Durasi Tidur:</b> %{x}<br>' +
    '<b>Jumlah:</b> %{z}<extra></extra>'
)

fig.show()
```

### **INTERPRETASI:**

* Mayoritas responden tidur 5–8 jam per hari.
* Tingkat nyeri sedang hingga tinggi paling banyak ditemukan pada durasi tidur 5–6 jam.
* Tidur yang kurang optimal kemungkinan berhubungan dengan meningkatnya keluhan nyeri.

### **Penjelasan**

Heatmap menunjukkan bahwa sebagian besar responden memiliki durasi tidur **5–6 jam** dan **7–8 jam**. Pada kedua kategori ini, tingkat nyeri yang paling dominan adalah tingkat 3 dan 4.

Kelompok dengan durasi tidur **5–6 jam** memiliki jumlah responden nyeri tingkat 3 paling tinggi. Hal ini menunjukkan bahwa durasi tidur yang lebih pendek mungkin berkaitan dengan meningkatnya rasa nyeri.

Sementara itu, responden dengan durasi tidur lebih panjang (lebih dari 8 jam) jumlahnya sangat sedikit sehingga belum dapat menggambarkan pola yang kuat.

### **2. — HUBUNGAN KUALITAS TIDUR DAN TINGKAT NYERI**

Grafik digunakan untuk melihat hubungan antara kualitas tidur dan tingkat nyeri menstruasi.

```python
import pandas as pd
import plotly.express as px

# Rapihin semua nama kolom
df.columns = df.columns.str.strip()

# Rename kolom
df = df.rename(columns={
    '14. Seberapa baik kualitas tidur Anda selama menstruasi?': 'kualitas_tidur',
    '4. Seberapa tinggi tingkat nyeri yang Anda rasakan saat menstruasi?': 'nyeri'
})

# Crosstab
crosstab_kualitas_tidur = pd.crosstab(
    df['kualitas_tidur'],
    df['nyeri']
)

# Ubah ke long format
bubble_data = crosstab_kualitas_tidur.reset_index().melt(
    id_vars='kualitas_tidur',
    var_name='Tingkat Nyeri',
    value_name='Jumlah Responden'
)

# Bubble chart
fig = px.scatter(
    bubble_data,
    x='kualitas_tidur',
    y='Tingkat Nyeri',
    size='Jumlah Responden',
    color='Jumlah Responden',
    text='Jumlah Responden',
    size_max=70,
    color_continuous_scale='Blues',
    title='Hubungan Kualitas Tidur dan Tingkat Nyeri'
)

# Layout
fig.update_layout(
    template='plotly_white',
    title_x=0.5,
    height=600,
    font=dict(size=14),
    xaxis_title='Kualitas Tidur',
    yaxis_title='Tingkat Nyeri'
)

# Hover
fig.update_traces(
    textposition='top center',
    hovertemplate=
    '<b>Kualitas Tidur:</b> %{x}<br>' +
    '<b>Tingkat Nyeri:</b> %{y}<br>' +
    '<b>Jumlah:</b> %{marker.size}<extra></extra>'
)

fig.show()
```

### **INTERPRETASI:**

* Mayoritas responden memiliki kualitas tidur sedang.
* Tingkat nyeri sedang (3) mendominasi hampir seluruh kategori kualitas tidur.
* Kualitas tidur yang kurang baik cenderung berkaitan dengan munculnya tingkat nyeri lebih tinggi.
* Tidur yang lebih berkualitas berpotensi membantu mengurangi keluhan nyeri.

### **Penjelasan**

Pada grafik bubble chart terlihat bahwa kualitas tidur **nilai 3** merupakan kategori yang paling banyak dimiliki responden. Pada kategori ini, tingkat nyeri yang paling dominan adalah **tingkat 3**, ditunjukkan oleh ukuran gelembung paling besar.

Responden dengan kualitas tidur lebih baik (4–5) masih mengalami nyeri, tetapi jumlahnya relatif lebih kecil dibanding kualitas tidur sedang. Sedangkan kualitas tidur rendah (1–2) juga menunjukkan adanya tingkat nyeri sedang hingga tinggi.
